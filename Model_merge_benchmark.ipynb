```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load base model and tokenizer
base_model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
base_model = AutoModelForCausalLM.from_pretrained(base_model_name, torch_dtype=torch.float16)

# Load fine-tuned models
math_code_model = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama_v1.1_math_code", torch_dtype=torch.float16)
finance_model = AutoModelForCausalLM.from_pretrained("ritamsharma/tinyllama-finance-v3", torch_dtype=torch.float16)
```

```python
def get_task_vector(fine_tuned_model, base_model):
    task_vector = {}
    print(fine_tuned_model.named_parameters(),'\n\n')
    for name, param in fine_tuned_model.named_parameters():
        if name in base_model.state_dict():
            task_vector[name] = param.data - base_model.state_dict()[name]
    # print(task_vector)
    return task_vector

math_code_vector = get_task_vector(math_code_model, base_model)
finance_vector = get_task_vector(finance_model, base_model)
```

```python
def merge_task_vectors(base_model, vector1, vector2, alpha=0.5):
    merged_state_dict = base_model.state_dict()
    for name in merged_state_dict:
        if name in vector1 and name in vector2:
            merged_state_dict[name] = merged_state_dict[name] + alpha * vector1[name] + (1 - alpha) * vector2[name]
    base_model.load_state_dict(merged_state_dict)
    return base_model

# Merge with equal weighting
merged_model = merge_task_vectors(base_model, math_code_vector, finance_vector, alpha=0.5)
```

```python
def merge_task_vectors_advanced(base_model, vector1, vector2, layer_scales={'': (0.5, 0.5)}):
    merged_state_dict = base_model.state_dict()
    for name, param in merged_state_dict.items():
        if name in vector1 and name in vector2:
            scale1 = 0
            scale2 = 0
            for layer_pattern, scales in layer_scales.items():
                if layer_pattern in name:
                    scale1, scale2 = scales
                    break

            merged_state_dict[name] = param + scale1 * vector1[name] + scale2 * vector2[name]
    base_model.load_state_dict(merged_state_dict)
    return base_model

# Example of advanced merging:
layer_specific_scaling = {
    'embed': (1.0, 1.0),
    'attn': (0.6, 0.4),
    'mlp': (0.4, 0.6),
    'lm_head': (0.2, 0.8),
    '': (0.5, 0.5) # Default scales
}

base_model_advanced = AutoModelForCausalLM.from_pretrained(base_model_name, torch_dtype=torch.float16) # Reload base model
math_code_vector = get_task_vector(math_code_model, base_model_advanced)
finance_vector = get_task_vector(finance_model, base_model_advanced)
merged_model_advanced = merge_task_vectors_advanced(base_model_advanced, math_code_vector, finance_vector, layer_specific_scaling)
```
