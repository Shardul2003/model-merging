````python
cloned_repo_name = "MergeLM"
current_dir = os.getcwd()
cloned_repo_path = os.path.join(current_dir, cloned_repo_name)

# Destination paths
# 'merging_methods' will be moved to the current directory
dest_merging_methods_path = os.path.join(current_dir, "model_merging_methods")
# 'utils' will be moved into the new 'merging_methods' directory
# So, its final path will be os.path.join(dest_merging_methods_path, "utils")

def main():
    print(f"Assuming cloned repository is at: {cloned_repo_path}")
    if not os.path.isdir(cloned_repo_path):
        print(f"Error: Directory '{cloned_repo_path}' not found.")
        print("Please ensure the 'MergeLM' folder is cloned and the path is correct.")
        return

    # Define source paths based on the cloned repository structure
    src_merging_methods = os.path.join(cloned_repo_path, "model_merging_methods")
    src_utils = os.path.join(cloned_repo_path, "utils")

    # 1. Move the merging_methods folder
    print(f"Attempting to move '{src_merging_methods}' to '{dest_merging_methods_path}'...")
    if os.path.exists(src_merging_methods):
        if os.path.exists(dest_merging_methods_path):
            print(f"Destination '{dest_merging_methods_path}' already exists. Removing it.")
            try:
                shutil.rmtree(dest_merging_methods_path)
                print(f"Removed existing directory: {dest_merging_methods_path}")
            except OSError as e:
                print(f"Error removing existing destination '{dest_merging_methods_path}': {e}")
                print("Please resolve this manually and try again.")
                return # Stop if we can't clear the destination
        try:
            shutil.move(src_merging_methods, dest_merging_methods_path)
            print(f"Successfully moved 'model_merging_methods' to '{dest_merging_methods_path}'.")
        except Exception as e:
            print(f"Error moving '{src_merging_methods}' to '{dest_merging_methods_path}': {e}")
            print("Please check paths and permissions.")
            return
    else:
        print(f"Error: Source folder '{src_merging_methods}' not found in '{cloned_repo_path}'.")
        print("Please check the structure of your cloned MergeLM repository.")
        return
    print("-" * 30)

    # 2. Move the utils folder into the new merging_methods folder
    dest_utils_final_path = os.path.join(dest_merging_methods_path, "utils")
    print(f"Attempting to move '{src_utils}' to '{dest_utils_final_path}'...")

    if os.path.exists(src_utils):
        # Ensure the parent directory for utils (i.e., dest_merging_methods_path) exists
        # This should be true if the previous step was successful
        if not os.path.exists(dest_merging_methods_path):
            print(f"Error: The target directory '{dest_merging_methods_path}' for utils does not exist. 'merging_methods' move might have failed.")
            return

        # If the final 'utils' path already exists inside 'merging_methods', remove it first
        if os.path.exists(dest_utils_final_path):
            print(f"Destination '{dest_utils_final_path}' already exists. Removing it.")
            try:
                shutil.rmtree(dest_utils_final_path)
                print(f"Removed existing directory: {dest_utils_final_path}")
            except OSError as e:
                print(f"Error removing existing destination '{dest_utils_final_path}': {e}")
                print("Please resolve this manually and try again.")
                return # Stop if we can't clear the destination

        try:
            # Move src_utils into dest_merging_methods_path.
            # shutil.move will rename src_utils to 'utils' inside dest_merging_methods_path if not already named 'utils'
            # or move the folder 'utils' into dest_merging_methods_path.
            shutil.move(src_utils, dest_merging_methods_path)
            # The above line moves the 'utils' folder from MergeLM/utils to ./merging_methods/utils
            print(f"Successfully moved 'utils' from '{src_utils}' into '{dest_merging_methods_path}', resulting in '{dest_utils_final_path}'.")
        except Exception as e:
            print(f"Error moving '{src_utils}' into '{dest_merging_methods_path}': {e}")
            print("Please check paths and permissions.")
            return
    else:
        print(f"Error: Source folder '{src_utils}' not found in '{cloned_repo_path}'.")
        print("Please check the structure of your cloned MergeLM repository.")
        return
    print("-" * 30)

    print("Script finished.")
    print(f"Check your current directory ('{current_dir}') for the 'merging_methods' folder.")
    print(f"The 'utils' folder should now be inside '{dest_merging_methods_path}'.")

main()
````

````python
import json
import os

def modify_adapter_target_modules(adapter_dir_path, new_target_modules):
    """
    Loads an adapter_config.json file, modifies its target_modules,
    and saves it back.

    Args:
        adapter_dir_path (str): Path to the directory containing the adapter_config.json.
        new_target_modules (list): A list of strings for the new target_modules.
    """
    config_path = os.path.join(adapter_dir_path, "adapter_config.json")

    if not os.path.exists(config_path):
        print(f"Error: adapter_config.json not found in {adapter_dir_path}")
        return False

    try:
        # Read the existing config
        with open(config_path, 'r') as f:
            adapter_config = json.load(f)
        print(f"Read adapter_config.json from {adapter_dir_path}")

        # Modify the target_modules
        original_target_modules = adapter_config.get("target_modules")
        adapter_config["target_modules"] = new_target_modules
        print(f"  Original target_modules: {original_target_modules}")
        print(f"  New target_modules set to: {new_target_modules}")

        # Write the modified config back
        with open(config_path, 'w') as f:
            json.dump(adapter_config, f, indent=4) # indent for pretty printing
        print(f"Successfully updated adapter_config.json in {adapter_dir_path}")
        return True
    except Exception as e:
        print(f"Error processing {config_path}: {e}")
        return False

# --- Configuration for adapter modification ---
# Paths to your LOCAL adapter directories (where they were downloaded)
LOCAL_MEDICAL_ADAPTER_DIR = "./downloaded_adapters/falcon-1b-medical-qa-10k"
LOCAL_TOXICITY_ADAPTER_DIR = "./downloaded_adapters/falcon-1b-toxicity-qa-10k"

# The new target modules based on your base model's structure
NEW_TARGET_MODULES = ["q_proj", "k_proj", "v_proj"]

# Modify the medical adapter
print(f"Modifying Medical Adapter at: {LOCAL_MEDICAL_ADAPTER_DIR}")
if os.path.exists(LOCAL_MEDICAL_ADAPTER_DIR):
    modify_adapter_target_modules(LOCAL_MEDICAL_ADAPTER_DIR, NEW_TARGET_MODULES)
else:
    print(f"Directory not found: {LOCAL_MEDICAL_ADAPTER_DIR}. Please ensure adapters are downloaded.")
print("-" * 30)

# Modify the toxicity adapter
print(f"Modifying Toxicity Adapter at: {LOCAL_TOXICITY_ADAPTER_DIR}")
if os.path.exists(LOCAL_TOXICITY_ADAPTER_DIR):
    modify_adapter_target_modules(LOCAL_TOXICITY_ADAPTER_DIR, NEW_TARGET_MODULES)
else:
    print(f"Directory not found: {LOCAL_TOXICITY_ADAPTER_DIR}. Please ensure adapters are downloaded.")
print("-" * 30)

print("Programmatic modification of adapter configs complete.")
print("You can now proceed with loading these adapters using their local paths in your main script.")

# --- Now, your main script that loads models using PeftModel would follow ---
# Ensure it uses LOCAL_MEDICAL_ADAPTER_DIR and LOCAL_TOXICITY_ADAPTER_DIR
# when calling PeftModel.from_pretrained()
````

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel # For loading adapters
import copy
import os
import re # For regex in exclude_param_names_regex
from collections import OrderedDict # For creating task vectors

# Attempt to import MergingMethod from your local file

print("Could not import MergingMethod from merging_methods.py. Using placeholder class.")
USING_EXTERNAL_MERGING_METHOD = False
# --- Placeholder for MergingMethod class if not in a separate file ---
class MergingMethod:
    def __init__(self, method_type):
        self.method_type = method_type
        print(f"Initialized MergingMethod (Placeholder) with type: {self.method_type}")

    def average_merging(self, models_to_merge, exclude_param_names_regex=None):
        if not models_to_merge:
            print("No models provided for average merging.")
            return {}
        print(f"Performing average merging on {len(models_to_merge)} models.")
        if exclude_param_names_regex is None:
            exclude_param_names_regex = []

        template_state_dict = models_to_merge[0].state_dict()
        averaged_params = {key: torch.zeros_like(param, dtype=torch.float32)
                            for key, param in template_state_dict.items()}
        num_models = len(models_to_merge)

        for model_idx, model in enumerate(models_to_merge):
            # print(f"  Processing model {model_idx+1}/{num_models} for averaging...")
            current_state_dict = model.state_dict()
            for key in averaged_params:
                if any(re.match(pattern, key) for pattern in exclude_param_names_regex):
                    if model_idx == 0:
                          averaged_params[key] = current_state_dict[key].clone().to(template_state_dict[key].dtype)
                    continue
                if key in current_state_dict:
                    averaged_params[key] += current_state_dict[key].to(torch.float32)
        for key in averaged_params:
            if any(re.match(pattern, key) for pattern in exclude_param_names_regex):
                continue
            if key in template_state_dict:
                averaged_params[key] = (averaged_params[key] / num_models).to(template_state_dict[key].dtype)
        print("Average merging calculation complete.")
        return averaged_params

    # Updated placeholder to match MergeLM library's expected signature for task_arithmetic
    def task_arithmetic(self, base_model_state_dict: OrderedDict,
                          task_vectors_state_dict: list[OrderedDict], # List of pre-computed task vector state_dicts
                          scaling_coefficient: float,
                          exclude_param_names_regex: list[str] = None):
        print(f"Performing task arithmetic (Placeholder) with {len(task_vectors_state_dict)} pre-computed task vectors and scaling: {scaling_coefficient}")
        if exclude_param_names_regex is None:
            exclude_param_names_regex = []

        # Start with a copy of the base model's parameters
        merged_params = OrderedDict((key, param.clone().to(torch.float32)) for key, param in base_model_state_dict.items())

        for tv_idx, task_vector_sd in enumerate(task_vectors_state_dict):
            # print(f"  Adding task vector {tv_idx+1}...")
            for key in merged_params:
                if any(re.match(pattern, key) for pattern in exclude_param_names_regex):
                    continue
                if key in task_vector_sd:
                    merged_params[key] += scaling_coefficient * task_vector_sd[key].to(torch.float32)

        for key in merged_params:
            if key in base_model_state_dict: # Cast back to original dtypes
                merged_params[key] = merged_params[key].to(base_model_state_dict[key].dtype)
        print("Task arithmetic (Placeholder) calculation complete.")
        return merged_params

    def copy_params_to_model(self, params_state_dict, model_instance):
        model_instance.load_state_dict(params_state_dict)
        print(f"Parameters copied to model instance: {model_instance.__class__.__name__}")


# --- Helper function to compute task vector state dicts ---
def compute_task_vector(model_state_dict: OrderedDict, base_model_state_dict: OrderedDict, exclude_regex: list = None) -> OrderedDict:
    """Computes a task vector: model_state_dict - base_model_state_dict."""
    if exclude_regex is None:
        exclude_regex = []
    task_vector_sd = OrderedDict()
    for key in base_model_state_dict:
        if key in model_state_dict:
            if any(re.match(pattern, key) for pattern in exclude_regex):
                # For excluded parameters, the task vector component is zero
                task_vector_sd[key] = torch.zeros_like(base_model_state_dict[key])
            else:
                task_vector_sd[key] = model_state_dict[key].to(dtype=base_model_state_dict[key].dtype) - base_model_state_dict[key]
        else:
            print(f"Warning: Key '{key}' not found in model_state_dict during task vector computation. Setting component to zero.")
            task_vector_sd[key] = torch.zeros_like(base_model_state_dict[key])
    return task_vector_sd

# --- Configuration ---
BASE_MODEL_ID = 'tiiuae/Falcon3-1B-Base'
LOCAL_MEDICAL_ADAPTER_DIR = "./downloaded_adapters/falcon-1b-medical-qa-10k"
LOCAL_TOXICITY_ADAPTER_DIR = "./downloaded_adapters/falcon-1b-toxicity-qa-10k"

FULL_MEDICAL_MODEL_DIR = "./full_medical_model_merged_adapters"
FULL_TOXICITY_MODEL_DIR = "./full_toxicity_model_merged_adapters"
BASE_MODEL_SAVE_DIR = "./base_model_original"

OUTPUT_MERGED_MODELS_PARENT_DIR = "./output_merged_models"
OUTPUT_DIR_BASE_med_tox_AVG = os.path.join(OUTPUT_MERGED_MODELS_PARENT_DIR, "base_med_tox_avg")
OUTPUT_DIR_BASE_med_tox_TASK = os.path.join(OUTPUT_MERGED_MODELS_PARENT_DIR, "base_med_tox_task_arithmetic")
OUTPUT_DIR_med_tox_AVG = os.path.join(OUTPUT_MERGED_MODELS_PARENT_DIR, "med_tox_avg")
OUTPUT_DIR_med_tox_TASK = os.path.join(OUTPUT_MERGED_MODELS_PARENT_DIR, "med_tox_task_arithmetic_no_base_avg")

TASK_ARITHMETIC_SCALING = 0.5
EXCLUDE_REGEX = []

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DTYPE = torch.float16

def save_model(model, tokenizer, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    model.save_pretrained(output_dir)
    if tokenizer: tokenizer.save_pretrained(output_dir)
    print(f"Model saved to: {output_dir}")

print("Loading models...")
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL_ID)
base_model_load_args = {"torch_dtype": DTYPE, "low_cpu_mem_usage": True, "trust_remote_code": True}

print(f"Loading Base Model: {BASE_MODEL_ID} for reference and saving...")
base_model_for_saving_and_ref = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args).to(DEVICE)
save_model(base_model_for_saving_and_ref, tokenizer, BASE_MODEL_SAVE_DIR)
print("-" * 30)

print(f"Loading and merging Medical Adapters from local path: {LOCAL_MEDICAL_ADAPTER_DIR}")
medical_base_for_peft = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args)
peft_medical_model = PeftModel.from_pretrained(medical_base_for_peft, LOCAL_MEDICAL_ADAPTER_DIR)
medical_model_merged_adapters = peft_medical_model.merge_and_unload().to(DEVICE, dtype=DTYPE)
save_model(medical_model_merged_adapters, tokenizer, FULL_MEDICAL_MODEL_DIR)
del medical_base_for_peft, peft_medical_model
if torch.cuda.is_available(): torch.cuda.empty_cache()
print("-" * 30)

print(f"Loading and merging Toxicity Adapters from local path: {LOCAL_TOXICITY_ADAPTER_DIR}")
toxicity_base_for_peft = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args)
peft_toxicity_model = PeftModel.from_pretrained(toxicity_base_for_peft, LOCAL_TOXICITY_ADAPTER_DIR)
toxicity_model_merged_adapters = peft_toxicity_model.merge_and_unload().to(DEVICE, dtype=DTYPE)
save_model(toxicity_model_merged_adapters, tokenizer, FULL_TOXICITY_MODEL_DIR)
del toxicity_base_for_peft, peft_toxicity_model
if torch.cuda.is_available(): torch.cuda.empty_cache()
print("-" * 30)
print("All initial models loaded/created and saved.")
print("-" * 30)

# --- MERGING OPERATIONS ---
print("\nStarting Merge Combination 1: Base + Full Medical + Full Toxicity")
avg_merger_1 = MergingMethod("average_merging")
models_to_avg_merge_1 = [base_model_for_saving_and_ref, medical_model_merged_adapters, toxicity_model_merged_adapters]
averaged_params_1 = avg_merger_1.average_merging(models_to_merge=models_to_avg_merge_1, exclude_param_names_regex=EXCLUDE_REGEX)
merged_model_avg_1 = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args).to(DEVICE)
avg_merger_1.copy_params_to_model(params_state_dict=averaged_params_1, model_instance=merged_model_avg_1)
print("Average Merging (Base + Full Medical + Full Toxicity) complete.")
save_model(merged_model_avg_1, tokenizer, OUTPUT_DIR_BASE_med_tox_AVG)
del merged_model_avg_1, averaged_params_1
if torch.cuda.is_available(): torch.cuda.empty_cache()

print("\nPerforming Task Arithmetic (Base + Medical_Delta_from_Base + Toxicity_Delta_from_Base)...")
task_merger_1 = MergingMethod("task_arithmetic")
base_s_dict_1b = base_model_for_saving_and_ref.state_dict()
task_vectors_s_dicts_1b = []
for model_obj in [medical_model_merged_adapters, toxicity_model_merged_adapters]:
    tv_s_dict = compute_task_vector(model_obj.state_dict(), base_s_dict_1b, EXCLUDE_REGEX)
    task_vectors_s_dicts_1b.append(tv_s_dict)

# The task_arithmetic method (from MergeLM or placeholder) applies these to the base_s_dict_1b
# and returns the new state_dict for the merged model.
task_merged_params_1 = task_merger_1.task_arithmetic(
    base_model_state_dict=base_s_dict_1b,
    task_vectors_state_dict=task_vectors_s_dicts_1b,
    scaling_coefficient=TASK_ARITHMETIC_SCALING,
    exclude_param_names_regex=EXCLUDE_REGEX
)
merged_model_task_1_result = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args).to(DEVICE)
task_merger_1.copy_params_to_model(params_state_dict=task_merged_params_1, model_instance=merged_model_task_1_result)
print("Task Arithmetic (Base + Medical_Delta + Toxicity_Delta) complete.")
save_model(merged_model_task_1_result, tokenizer, OUTPUT_DIR_BASE_med_tox_TASK)
del merged_model_task_1_result, task_merged_params_1
if torch.cuda.is_available(): torch.cuda.empty_cache()
print("-" * 30)

print("\nStarting Merge Combination 2: Full Medical + Full Toxicity")
avg_merger_2 = MergingMethod("average_merging")
models_to_avg_merge_2 = [medical_model_merged_adapters, toxicity_model_merged_adapters]
averaged_params_2 = avg_merger_2.average_merging(models_to_merge=models_to_avg_merge_2, exclude_param_names_regex=EXCLUDE_REGEX)
merged_model_avg_2 = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args).to(DEVICE)
avg_merger_2.copy_params_to_model(params_state_dict=averaged_params_2, model_instance=merged_model_avg_2)
print("Average Merging (Full Medical + Full Toxicity) complete.")
save_model(merged_model_avg_2, tokenizer, OUTPUT_DIR_med_tox_AVG)
del merged_model_avg_2, averaged_params_2
if torch.cuda.is_available(): torch.cuda.empty_cache()

print("\nPerforming Task Arithmetic (Medical_Model + scaled * (Toxicity_Model - Medical_Model))...")
task_merger_2b = MergingMethod("task_arithmetic")
base_s_dict_2b = medical_model_merged_adapters.state_dict() # Medical model is base for this operation
task_vectors_s_dicts_2b = [compute_task_vector(toxicity_model_merged_adapters.state_dict(), base_s_dict_2b, EXCLUDE_REGEX)]

task_merged_params_2b = task_merger_2b.task_arithmetic(
    base_model_state_dict=base_s_dict_2b,
    task_vectors_state_dict=task_vectors_s_dicts_2b,
    scaling_coefficient=TASK_ARITHMETIC_SCALING,
    exclude_param_names_regex=EXCLUDE_REGEX
)
merged_model_task_2b_result = AutoModelForCausalLM.from_pretrained(BASE_MODEL_ID, **base_model_load_args).to(DEVICE)
task_merger_2b.copy_params_to_model(params_state_dict=task_merged_params_2b, model_instance=merged_model_task_2b_result)
print("Task Arithmetic (Medical + scaled*(Toxicity - Medical)) complete.")
save_model(merged_model_task_2b_result, tokenizer, OUTPUT_DIR_med_tox_TASK)
del merged_model_task_2b_result, task_merged_params_2b
if torch.cuda.is_available(): torch.cuda.empty_cache()
print("-" * 30)

print("All merging tasks finished.")

````

````python
def move_subfolders_to_content(source_parent_dir_name="merged_models", destination_root_dir="/content/"):
    """
    Moves all subdirectories from a specified source parent directory
    to a destination root directory (e.g., /content/).

    Args:
        source_parent_dir_name (str): The name of the folder containing the subdirectories to move.
                                      Assumed to be in the current working directory.
        destination_root_dir (str): The absolute path to the destination root directory.
    """
    current_working_dir = os.getcwd()
    source_parent_full_path = os.path.join(current_working_dir, source_parent_dir_name)

    print(f"Source directory: {source_parent_full_path}")
    print(f"Destination root: {destination_root_dir}")
    print("-" * 30)

    if not os.path.isdir(source_parent_full_path):
        print(f"Error: Source directory '{source_parent_full_path}' not found.")
        print("Please ensure the 'merged_models' directory exists in your current working directory.")
        return

    # List all items (files and folders) in the source_parent_full_path
    items_in_source_parent = os.listdir(source_parent_full_path)

    if not items_in_source_parent:
        print(f"No items found in '{source_parent_full_path}'. Nothing to move.")
        return

    moved_items_count = 0
    for item_name in items_in_source_parent:
        source_item_path = os.path.join(source_parent_full_path, item_name)

        # Check if the item is a directory
        if os.path.isdir(source_item_path):
            destination_item_path = os.path.join(destination_root_dir, item_name)
            print(f"Processing directory: '{item_name}'")
            print(f"  Source: {source_item_path}")
            print(f"  Destination: {destination_item_path}")

            # Check if the destination already exists
            if os.path.exists(destination_item_path):
                print(f"  Warning: Destination '{destination_item_path}' already exists.")
                # Option 1: Skip
                # print(f"  Skipping '{item_name}' as destination already exists.")
                # continue

                # Option 2: Remove existing and replace (use with caution)
                print(f"  Removing existing destination '{destination_item_path}' to replace it.")
                try:
                    shutil.rmtree(destination_item_path)
                    print(f"  Successfully removed existing '{destination_item_path}'.")
                except OSError as e:
                    print(f"  Error removing existing '{destination_item_path}': {e}")
                    print(f"  Skipping move for '{item_name}'. Please resolve manually.")
                    continue # Skip to the next item

            # Move the directory
            try:
                shutil.move(source_item_path, destination_item_path)
                print(f"  Successfully moved '{item_name}' to '{destination_item_path}'.")
                moved_items_count += 1
            except Exception as e:
                print(f"  Error moving '{item_name}': {e}")
                print(f"  Please check paths and permissions for '{source_item_path}' and '{destination_item_path}'.")
            print("-" * 10)
        else:
            print(f"Skipping non-directory item: '{item_name}' at '{source_item_path}'")
            print("-" * 10)

    print("-" * 30)
    if moved_items_count > 0:
        print(f"Finished moving {moved_items_count} director(y/ies).")
        print(f"The original '{source_parent_full_path}' directory might now be empty or only contain files.")
    else:
        print("No directories were moved. Check if 'merged_models' contained any subdirectories.")


move_subfolders_to_content(source_parent_dir_name="output_merged_models",destination_root_dir="/content/")
````
